#* PREPROCESSING*#

```python
meanl=df['LotFrontage'].mean(skipna=True)
```

```python
meanl
```

```python
df['LotFrontage'].fillna(meanl,inplace=True)
```

```python
for i in df.isnull().sum():
  print(i)
```

```python

null_counts = df.isnull().sum()
selected_columns = null_counts.loc[null_counts > 600]
selected_index = selected_columns.index
print(selected_columns)
print(selected_index)
```

```python
selected_index=selected_index.tolist()
```

```python
selected_index
```

```python
a=[]
for i in range(0,len(df.columns)):
  if(df.columns[i] in selected_index):
    a.append(i)
```

```python
a
```

```python
df.drop(df.columns[a],axis=1,inplace=True)
```

```python
df.isnull().any()
```

```python

for i in df.isnull().sum():
  print(i)
```

```python
for i in df.columns:
  if(df[i].dtype!='object'):
    meanl=df[i].mean(skipna=True)
    df[i].fillna(meanl,inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
```

```python
for i in df.columns:
  if(df[i].dtype=='object'):
    df[i]=label_encoder.fit_transform(df[i])
```

```python
for i in df.columns:
  if(df[i].dtype=='object'):
    df[i].fillna(method='ffill')
```

```python
df=df.fillna(method='bfill')
```

```python
x=df.drop(['Id','SalePrice'],axis=1)
y=df['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20)
```

```python
from sklearn.linear_model import LinearRegression
b=LinearRegression()
b.fit(x_train,y_train)
y_pred1=b.predict(x_test)
```

```python
from sklearn.metrics import r2_score
ac=r2_score(y_pred1,y_test)
```
